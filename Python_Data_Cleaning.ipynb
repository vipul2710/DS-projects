```python
regions = ["US","U.S.A.","usa","UK","Uk","IN","Ind","in","uSa"]
```

```python
for i in range(9,30):
    data.append({
        "id": str(100+i%10),
        "date": random.choice(["2023-09-10","10-09-2023","2023/09/10","Sep 10 2023"]),
        "customer_name": random.choice(["Ann","Ben "," Carl","dave","eva"]),
        "region": random.choice(regions),
        "product": random.choice(["A","B","C", None]),
        "sales_amount": random.choice(["250","NA","", str(random.randint(50,900)), "-10"])
    })
```

```python
df_messy = pd.DataFrame(data)
df_messy.to_csv("sales_data_messy.csv", index=False)
print(f"✅ Dataset created: sales_data_messy.csv  | Rows: {len(df_messy)}")
```

```python
df['region'].value_counts().head(10)
```

```python
df['id'] = df['id'].astype(str)
```

```python
#keep first occurence and drop rest
df = df.drop_duplicates(subset='id', keep='first')
```

```python
df = df.dropna(subset=['id'])
df
```

```python
#date conversion
df['date'] = pd.to_datetime(df['date'], errors='coerce', dayfirst=False)
```

```python
df['date'].isna().sum()
```

```python
df['customer_name'] = df['customer_name'].str.strip().str.title()
```

```python
df['customer_name'].unique()
```

```python
df['region'] = df['region'].astype(str).str.lower().str.strip()
```

```python
#removing puncutation
df['region'] = df['region'].str.replace(r'[^a-z]', '', regex=True)
```

```python
region_map = {
    r'^(us|usa|us a|usaa)$': 'usa',
    r'^(uk|uuk|unit?edkingdom)$': 'uk',   # permissive, but mapping regex anchors help
    r'^(in|ind|india)$': 'india'
}
```

```python
for pat, val in region_map.items():
    df['region'] = df['region'].replace(pat, val, regex=True)
```

```python
print("Unique regions after mapping:", df['region'].unique())
```

```python
df['region'] = df['region'].astype('category')
```

```python
print(df['region'].value_counts(dropna=False))
```

```python
df['product'].value_counts().head(10)
```

```python
df['product'] = df.groupby('region')['product'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else None)
)
```

```python
df['product'].isna().sum()
df['product'].value_counts()
```

```python
df['sales_amount'] = df['sales_amount'].replace(['NA', '', 'None'], pd.NA)
```

```python
df['sales_amount'] = pd.to_numeric(df['sales_amount'], errors='coerce')
```

```python
df.loc[df['sales_amount'] < 0, 'sales_amount'] = pd.NA
```

```python
median_sales = df['sales_amount'].median()
df['sales_amount'] = df['sales_amount'].fillna(median_sales)
```

```python
print(df['sales_amount'].describe())
print("Missing values:", df['sales_amount'].isna().sum())
print("Negative values:", (df['sales_amount'] < 0).sum())
```

```python
df.to_csv("sales_data_cleaned.csv", index=False)
print("✅ Cleaned dataset saved as 'sales_data_cleaned.csv'")
```
